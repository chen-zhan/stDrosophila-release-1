# **Reconstruction 3D model based on spatial transcriptomics**

## **Import packages**

```python
import warnings
warnings.filterwarnings("ignore")
from stDrosophila.pl import *
```

## **Load data**

```python
ipath = "D:\BGIpy37_pytorch113\Test_Data\E16-18h_a_all_anno_v1.h5ad"
adata = ad.read(ipath)
del adata.var
del adata.obsm
del adata.layers
adata.obs = adata.obs[["slice_ID", "new_x", "new_y", "new_z", "anno"]]
adata.obs.columns = ["slice", "x", "y", "z", "anno"]
adata.obs["x"] = adata.obs["x"] - adata.obs["x"].min()
adata.obs["x"] = adata.obs["x"].round().astype(float)
adata.obs["y"] = adata.obs["y"] - adata.obs["y"].min()
adata.obs["y"] = adata.obs["y"].round().astype(float)
adata.obs["z"] = adata.obs["z"].astype(float)
adata.obsm["spatial"] = adata.obs[["x", "y"]].values
print(adata)
print(adata.obs)
print(adata.obsm["spatial"])
```

## **Create an unstructured mesh**

- **Create an points cloud (pyvista.PolyData)**

```python
mesh, surface = build_3Dmodel(adata=adata, coordsby=["x", "y", "z"], groupby="anno", group_show="fat body",
                              gene_show="all", colormap=["dodgerblue"], alphamap=1.0, mask_color="whitesmoke",
                              mask_alpha=0.2, smoothing=True, voxelize=False, unstructure=True)
easy_plot(surface=surface, meshes=mesh, scalars="anno_rgba", shape=(1, 1), cpos="iso")
```
